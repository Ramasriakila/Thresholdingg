```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
road_image = cv2.imread('cub.jpg')
```

```python
color_image = cv2.cvtColor(road_image, cv2.COLOR_BGR2RGB)
```

```python
gray_image = cv2.imread('cub.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
reduce_noise = cv2.GaussianBlur(gray_image, (3,3), 0)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(road_image[:,:,::-1])
plt.title("Original Image")
plt.axis("off")
```

```python
plt.imshow(gray_image, cmap='gray')
plt.title("Gray Image")
plt.axis("off")
plt.show()
```

```python
canny = cv2.Canny(gray_image, 120, 150)
```

```python
plt.imshow(canny, cmap='gray')
plt.title("Canny Edge Detector")
plt.axis("off")
plt.show()
```

```python
lines = cv2.HoughLinesP(canny, 1, np.pi/180, threshold=80, minLineLength=50, maxLineGap=250)
```

```python
if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(road_image, (x1, y1), (x2, y2), (255, 0, 0), 3)
```

```python
plt.imshow(road_image)
plt.title("Result Image with Hough Lines")
plt.axis("off")
plt.show()
```
